```python
# matplotlib - library used for graphics
# for matplotlib video tutorials: https://www.youtube.com/watch?v=UO98lJQ3QGI&list=PL-osiE80TeTvipOqomVEeZ1HRrcEvtZB_
# matplotlib doc tutorial: https://www.datacamp.com/community/tutorials/matplotlib-tutorial-python
from matplotlib import pyplot as plt
xValues = [10,20,30]
yValues = [100,120,110]
plt.plot(xValues,yValues)
plt.show()
```

```python
# we can set titles ans x\y labels
plt.title('this is a nice title')
plt.xlabel('x values')
plt.ylabel('y values')
plt.plot(xValues,yValues)
plt.show()
```

```python
# let's do some practicle plot
# read the trees.csv file and display height vs volume graph
```

```python
import pandas as pd
treesData = pd.read_csv('trees.csv')
treesData.head()
```

```python
treesHeight = treesData['Height']
treesVolume = treesData['Volume']
```

```python
plt.plot(treesHeight,treesVolume,'bo')
plt.title('Trees Volume vs Height')
plt.ylabel('Volume')
plt.xlabel('Height')
plt.show()
```
